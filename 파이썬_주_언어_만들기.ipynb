## 미션 문제 - 1

```python
# 백준 1330
# 두 정수 A와 B가 주어졌을 때, A와 B를 비교하는 프로그램을 작성하시오.

a, b = map(int, input().split())

if a > b:
    print(">")
elif a < b:
    print("<")
elif a == b:
    print("==")
```

# 파이썬 사고방식 만들기 - Day 2

## for문이 필요한 이유?

```python
# 숫자 5개를 입력받아서 다 더해라
#요렇게 쓰면 효율이 떨어짐..

a = int(input())
b = int(input())
c = int(input())
d = int(input())
e = int(input())
```

## for문의 기본 구조(이거 진짜 중요)

```python
# for문 = 정해진 횟수만큼 같은 일을 시킴

for i in range(5): # 0,1,2,3,4
    print(i) # 지금 반복 중인 숫자
```

## input + for = 알고리즘 필수 조합

```python
# 숫자 5번을 입력받아서 하나씩 total에 누적해라

total = 0
for i in range(5):
    num = int(input())
    total += num
print(total)
```

## i는 그냥 번호표다

```python
# i는 순서가 필요할 떄
# 몇 번째인지 알고 싶을 때
# 인덱스 다룰때만 쓰임

for i in range(3):
    print(i)
```

## 오늘의 미션

```python
# 백준 2739
# N을 입력받은 뒤, 구구단 N단을 출력하는 프로그램을 작성하시오. 출력 형식에 맞춰서 출력하면 된다.

n = int(input())

for i in range(1, 10):
    print(n, "*", i, "=", n*i)
```

# 파이썬 사고방식 만들기 - Day 3

## 조건 + 반복 합체가 중요한 이유?

```python
# 문제 대부분은 이러한 구조 -> 여러 번 반복하면서, 특정 조건일때 뭔가 해라
# 이것이 for(여러 개 처리) + if 선별
```

## 가장 기본 합체 예제

```python
for i in range(1, 11): # 1부터 10까지 돌면서
    if i % 2 == 0: # 짝수면 출력해라
        print(i)
```

## 진짜 중요한 사고법(시험에도 나옴)

```python
# 문제 풀때 이렇게 나눠라
# 몇 번 반복 -> for
# 언제 행동 -> if
# 뭘 할 건데? -> print, +=, 저장
```

## 누적 + 조건 예제

```python
total = 0

for i in range(1, 101):
    if i % 2 == 0:
        total += i

print(total)
```

## 미션 문제(백준)

```python
# 백준 10871(X보다 작은 수)
# 정수 N개로 이루어진 수열 A와 정수 X가 주어진다.
# 이때, A에서 X보다 작은 수를 모두 출력하는 프로그램을 작성하시오.

n, x = map(int, input().split())

for i in range(n):
    a = map(int, input().split())
    if a < x:
        print(a)
```
