```python
# 파이썬 - 사람이 읽기 쉬운 언어
# 공학에서 쓰는 이유
# 빠른 프로토타이핑
# 알고리즘 연습
# AI / 자동화 / HCI 연구까지 가능
```

## 첫 코드 : print

```python
# 위 아래로 순서대로 실행
# print() = 화면에 보여주는 역할

rint("Hello, Python")
print(10)
print(10 + 3)
```

## 변수 = 값에 이름 붙이기

```python
# =은 같다 x
# =은 저장한다 o

a = 10
b = 3
print(a+b)
```

## 입력받기 - input

```python
# input()으로 받는 건 무조건 문자열(str)

a = input()
print(a)
```

```python
# 이렇게 씀
# int가 필요한 이유 - 문자열을 더하면 붙고, 숫자는 계산됨

a = int(input())
b = int(input())
print(a + b)
```

## 실전 미션

```python
##  백준 1000번(A+B)

a, b = map(int, input().split())

print(a+b)
```
