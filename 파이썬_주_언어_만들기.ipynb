## Python Day 2 - "조건문(IF")

## 왜 조건문이 필요할까?

```python
# 온도가 30도 이상 -> 에어컨 on
# 아니면 -> off
# 이걸 코드로 판단하게 하는게 if
```

## if 기본 구조(형태 익히기)

```python
if 조건:
    실행문
```

```python
# 들여쓰기 = 블록
# 파이썬은 {} 대신 공백으로 구분

x = 10

if x > 0:
    print("양수")
```

## 비교 연산자 정리

```python
# | 기호   | 의미  |
# | ---- | --- |
# | `>`  | 크다  |
# | `<`  | 작다  |
# | `==` | 같다  |
# | `!=` | 다르다 |
# | `>=` | 이상  |
# | `<=` | 이하  |
```

## if / elif / else 구조

```python
# 경우의 수를 나눈다는 사고가 중요함

if a > b:
    print("a가 크다")

elif a < b:
    print("b가 크다")

else:
    print("같다")
```

## 실전연결

```python
# 백준 1330번
# 두 정수 A, B | A와 B의 대소관계를 출력

a, b = map(int, input().split())

if a > b:
    print(">")

elif a < b:
    print("<")

else:
    print("==")
```
