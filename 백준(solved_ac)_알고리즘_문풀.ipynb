```python
# 백준 2752
# 제일 작은 수, 그 다음 수, 제일 큰 수를 차례대로 출력한다.

list1 = list(map(int,input().split()))

list1.sort()
print(*list1) ## *를 사용하면 인자값을 넘겨줘서 숫자 그대로의 값이 나옴
```

```python
# 백준 25640
# 진호와 MBTI 유형이 같은 사람의 수를 출력한다.

n = input()
num = int(input())
count = 0

for _ in range(num):
  if input() == n:
    count += 1

print(count)
```
