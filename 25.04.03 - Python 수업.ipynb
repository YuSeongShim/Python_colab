## 중첩 루프 사용하기

```python
# 반복문 안에 반복문이 들어가는 형태를 중첩 루프 라고 함
# 인덱스 변수는 index의 첫 글자를 따서 i를 사용하는데, 반복문 안에 반복문을 들어갈 때는 변수 i,j,k 처럼 순서대로 지음

for i in range(5):  # 5번 반복. 바깥쪽 루프는 세로 방향
    for j in range(5): # 5번 반복. 안쪽 루프는 가로 방향
        print('j:', j, sep=' ', end = ' ') # j값 출력
                                           # end에 ' '를 지정하여 줄 바꿈 대신 한 칸 띄움
    print('i:', i, '\\n', sep=' ')        # i값 출력, 개행 문자 모양도 출력
                                          # 가로 방향으로 숫자를 모두 출력한 뒤 다음 줄로 넘어감
                                          #(print는 기본적으로 출력 후 다음 줄로 넘어감)
```

## 사각형 별 출력하기

```python
for i in range(5):
    for j in range(5):
        print('*', end=' ')
    print()
```

## 사각형 모양 바꾸기

```python
for i in range(3): # 3번 반복. 세로 방향
    for j in range(7): # 5번 반복. 가로 방향
        print('*', end=' ')
    print()
```

## 계단식으로 별 출력하기

```python
for i in range(5):
    for j in range(5):
        if j <= i:  # 세로 방향 i변수 만큼
            print('*', end=' ')
    print()
```

## 대각선으로 별 출력하기

```python
for i in range(5):
    for j in range(5):
        if j == i:
            print('*', end=' ')
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j == i:
              print('*', end='')
        else:
              print(' ', end='')
    print()
```

## 퀴즈

```python
# 다음 코드에 대한 설명으로 잘못된 것을 모두 고르시오

for i in range(5):
    for j in range(5):
        print('*', end='')
    print()

# c, e
```

```python
# 다음과 같은 모양으로 별을 출력할 때 밑줄 부분에 들어가야 할 조건식을 고르세요

for i in range(5):
    for j in range(5):
        if j <= i:
            print('*', end='')
    print()

    # c
```

## 연습문제: 역삼각형 모양으로 출력하기

```python
for i in range(5):
    for j in range(5):
        if j < i:
            print(' ', end='')
        else:
            print('*', end='')
    print()
```

## 심사문제: 산 모양으로 별 출력하기

```python
for i in range(5):
    for j in range(5):
        if j > i:
            print(' ', end='')
        else:
            print('*', end='')
    print()
```

```python
n = int(input())

for i in range(n):
    for j in range(5):
        if j > i:
            print(' ', end='')
        else:
            print('*', end='')
    print()
```
